## Load Data from CSV

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('outputs/results.csv')
```

## Results Tables

```python
# Upewniamy się, że kolumny są w odpowiedniej kolejności
columns_order = ['ml_algorithm', 'fairness_mode', 'alpha_mode', 'alpha_value', 'test_accuracy', 'fairness_score']
```

```python
# Grupowanie i wyświetlanie
for ml in df['ml_algorithm'].unique():
    for fairness in df['fairness_mode'].unique():
        for alpha in df['alpha_mode'].unique():
            subset = df[
                (df['ml_algorithm'] == ml) &
                (df['fairness_mode'] == fairness) &
                (df['alpha_mode'] == alpha)
            ][columns_order]

            if not subset.empty:
                # Zaokrąglenie
                subset[['test_accuracy', 'fairness_score']] = subset[['test_accuracy', 'fairness_score']].round(3)

                print(f"\n=== ML Algorithm: {ml} | Fairness Mode: {fairness} | Alpha Mode: {alpha} ===")
                display(subset.sort_values(by='alpha_value')) 
```

## Line Charts - Accuracy vs Alpha & Fairness vs Alpha (Alpha = const)

```python
plt.figure(figsize=(10,6))
```

```python
unique_alpha_modes = df['alpha_mode'].unique()

for alpha_mode in unique_alpha_modes:
    # Tworzymy wykresy dla danego alpha_mode
    fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharex=True)
    
    # Filtrowanie danych dla danego alpha_mode
    df_alpha = df[df['alpha_mode'] == alpha_mode]
    
    # Accuracy
    for mode in df_alpha['fairness_mode'].unique():
        subset = df_alpha[df_alpha['fairness_mode'] == mode]
        axes[0].plot(subset['alpha_value'], subset['test_accuracy'], marker='o', label=mode)

    axes[0].set_title("Alpha vs Test Accuracy")
    axes[0].set_xlabel("Alpha")
    axes[0].set_ylabel("Accuracy")
    axes[0].legend(title="Fairness Mode", loc='lower right')
    axes[0].grid(True)
    axes[0].set_xlim(-0.02, 1.02)
    axes[0].set_ylim(0, 1)

    # Unfairness
    for mode in df_alpha['fairness_mode'].unique():
        subset = df_alpha[df_alpha['fairness_mode'] == mode]
        axes[1].plot(subset['alpha_value'], subset['fairness_score'], marker='o', label=mode)

    axes[1].set_title("Alpha vs Unfairness")
    axes[1].set_xlabel("Alpha")
    axes[1].set_ylabel("Unfairness")
    axes[1].legend(title="Fairness Mode", loc='upper left')
    axes[1].grid(True)
    axes[1].set_xlim(-0.02, 1.02)
    axes[1].set_ylim(-0.02, 0.5)

    # Tytuł dla całej figury
    plt.suptitle(f"Wyniki modeli — alpha_mode: {alpha_mode}", fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```
